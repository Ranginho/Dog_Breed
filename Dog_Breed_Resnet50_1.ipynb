```python
from google.colab import drive
drive.mount('/content/drive')

import numpy as np 
import pandas as pd
import os
import cv2
from tqdm import tqdm
from random import shuffle
import pickle
import random

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import RMSprop

IMG_SHAPE = 224

datagen = ImageDataGenerator(
    rescale=1./255.,
    rotation_range = 40,
    zoom_range = 0.2,
    horizontal_flip = True
    )

train_generator = datagen.flow_from_directory(
        'drive/MyDrive/dog-breed-identification/train_labeled',
        target_size=(IMG_SHAPE, IMG_SHAPE),
        batch_size=32,
        class_mode='categorical',
        seed=42,
        shuffle=True,)
        

validation_generator = datagen.flow_from_directory(
        'drive/MyDrive/dog-breed-identification/valid_folders',
        target_size=(IMG_SHAPE, IMG_SHAPE),
        batch_size=32,
        class_mode='categorical',
        seed=42,
        shuffle=True,) 
```

```python
resnet = tf.keras.applications.ResNet50V2(input_shape = [IMG_SHAPE,IMG_SHAPE,3], weights = 'imagenet', include_top = False)

for layer in resnet.layers:
    layer.trainable = False

x = tf.keras.layers.Flatten()(resnet.output)
pred = tf.keras.layers.Dense(120, activation='softmax')(x)
model = tf.keras.models.Model(inputs=resnet.input, outputs=pred)

opt = tf.keras.optimizers.Adam(learning_rate = 1e-5)
model.compile(optimizer=opt,loss='categorical_crossentropy',metrics=['accuracy'])
train_step = 200

model.summary()
```

```python
resnet50 = model.fit(train_generator,validation_data = validation_generator,epochs = 20,steps_per_epoch = train_step)
```
